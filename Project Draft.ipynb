# Splitting the Data into Train and Test


First we will devide the Data into Target and Predictor variables


```python
X = df.drop(columns=['Total Medicare reimbursements per enrollee (2014)'])
```

```python
X.head(2)
```

```python
y= df['Total Medicare reimbursements per enrollee (2014)']
```

```python
y.info()
```

### Dividing the Data as Training Set and Testing Set



```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25, random_state=42)
```

# Model 1: Liner Regression



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Initialize and train the Linear Regression model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
#print(y_pred)
```

### Predicting the Test Results

```python
# Just testing the model for one of the rows, below the prediction came at 9833 but the actual value is 9859
model.predict([[1001,4649.0,4053.0,2539.0,1949.0,477.0,628.0,233.0]])
```

### Evaluate the model

```python
from sklearn.metrics import r2_score

r2=r2_score(y_test,y_pred)
print(f"R2 score for test set is: {r2}")
```

```python
# The model has got r2 score of 0.9 which is a good score
```

### Plot the results

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted')
```

### Predicitng the Values

```python
pred_y_df=pd.DataFrame({'Actual value': y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
pred_y_df[0:10]
```

# Model 2: Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Creating the Decision Tree Regressor model
model2 = DecisionTreeRegressor(random_state=42)
```

```python
# Training the model
model2.fit(X_train, y_train)
```

```python
# Define the hyperparameters and their values to be tested
param_grid = {
    'max_depth': [None, 12, 10, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_features': [None, 'sqrt', 'log2']
}
```

```python
# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(estimator=model2, param_grid=param_grid, cv=5, n_jobs=-1, scoring='r2')
grid_search.fit(X_train, y_train)
```

```python
# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best Hyperparameters: {best_params}")
print(f"Best CV R2 Score: {best_score}")
```

```python
# Train the model with the best hyperparameters
best_model2 = grid_search.best_estimator_
best_model2.fit(X_train, y_train)
```

```python
# Making predictions
y_pred2 = model2.predict(X_test)
```

### Evaluate the Model

```python
from sklearn.metrics import r2_score

r2=r2_score(y_test,y_pred2)
print(f"R2 score for test set is: {r2}")
```

### Plot the results

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
plt.scatter(y_test,y_pred2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual Vs Predicted')
```
